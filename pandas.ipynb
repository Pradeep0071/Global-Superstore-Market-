```python
### 8. Find the overall rank of "India" where the 'Profit' is maximum under the category 'Technology'.
```

```python
# SIR WILL EXPLAIN

# Step 1: Filter data for Technology category
tech_df = df[df['Category'] == 'Technology']

# Step 2: Group by Country (or State if required), and sum the Profit
country_profit = tech_df.groupby('Country')['Profit'].sum()

# Step 3: Rank the countries based on Profit (descending = highest profit is rank 1)
ranked = country_profit.rank(ascending=False, method='max')

# Step 4: Get the rank of India
india_rank = ranked['India']
print(f"India's rank: {india_rank}")
```
